# Parsing a Map Image

### Step 1: read the file

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("map.png", cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap='gray') # why doesn't this work? 
```

### Step 2: Calculate Color Histogram

```python
import scipy.signal as sig
```

```python

bins = 256
hist = cv2.calcHist([image], [0], None, [bins], [0, bins])
peaks = sig.find_peaks(hist.flatten(), distance=24, threshold=100)[0]
# plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Bins")
plt.ylabel("# of Pixels")
plt.plot(hist)
plt.vlines(peaks[0], ymin=0, ymax=800000, color='red', label='RMSE')
plt.xlim([0, bins])
plt.show()
```

```python
mask = (image == peaks[0]).astype('uint8')
masked = cv2.bitwise_and(image, image, mask=mask)
plt.imshow(masked)
```

```python
mask = (image == peaks[1]).astype('uint8')
masked = cv2.bitwise_and(image, image, mask=mask)
plt.imshow(masked)
```
